```python
# Libraries and Dependencies
import pandas as pd
import numpy as np
```

```python
housing_data = pd.read_csv("Data/American_Housing_Data_20231209.csv")

housing_data.head()
```

```python
housing_data.isnull().sum()
```

```python
housing_data = housing_data.dropna()

housing_data.isnull().sum()
```

```python
housing_data.dtypes
```

```python
len(housing_data)
```

```python
dupes = housing_data.duplicated(subset=["Address"])

dupe_list = list(np.where(dupes)[0])

len(dupe_list)
```

```python
for index in dupe_list:
    print(housing_data.loc[index, "Address"])
```

```python
housing_data = housing_data.drop_duplicates(subset=['Address'])

len(housing_data)
```

```python
housing_data.loc[housing_data["City"] == housing_data["State"]]
```

```python
housing_data.loc[housing_data["City"] == "New York", "City"] = "New York City"

housing_data.head()
```

```python
housing_data.to_csv("Data/cleaned_American_Housing_Data_2023.csv", index=False)
```

```python
value_forecasts = pd.read_csv("Data/Zip_zhvf_growth_uc_sfrcondo_tier_0.33_0.67_month.csv")

value_forecasts.head()
```

```python
value_forecasts = value_forecasts.drop(columns=['RegionID', 'SizeRank', 'RegionType', 'StateName', 'State', 'Metro', 'City', 'CountyName'])

value_forecasts.head()
```

```python
value_forecasts = value_forecasts.rename(columns={"RegionName": "Zip Code",
                                                    "2024-04-30": "3-Month",
                                                     "2024-06-30": "6-Month",
                                                      "2025-03-31": "1-Year"})

value_forecasts.head()
```

```python
value_forecasts.isnull().sum()
```

```python
value_forecasts = value_forecasts.dropna()

value_forecasts.isnull().sum()
```

```python
value_forecasts.dtypes
```

```python
dupes = value_forecasts.duplicated(subset=["Zip Code"])

dupe_list = list(np.where(dupes)[0])

len(dupe_list)
```

```python
value_forecasts.to_csv("Data/Zillow_value_forecast.csv", index=False)
```

```python
housing_value_joined = housing_data.join(value_forecasts.set_index('Zip Code'), on="Zip Code", how='left')

housing_value_joined
```
